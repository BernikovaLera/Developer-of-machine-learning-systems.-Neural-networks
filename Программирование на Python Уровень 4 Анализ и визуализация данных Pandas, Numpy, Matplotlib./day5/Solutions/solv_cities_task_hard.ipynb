1. Найдите города, которые расположены в Центральном федеральном округе в радиусе 100 км от административного центра(Москва)с населением <= 50т. человек

```python
df['Регион'].unique()
```

```python
# Ищем строку, где Регион — Москва (не область, а сам город)
moscow_row = df[df['Регион'] == 'Москва'].iloc[0]

# Получаем координаты
moscow_lat = moscow_row['Широта']
moscow_lon = moscow_row['Долгота']

print(moscow_row, moscow_lat, moscow_lon)
```

```python
df.loc[(df['Регион'] == 'Москва') & (df['Город'].isna()), 'Город'] = 'Москва'
```

```python
df[df['Регион'] == 'Москва']
```

```python
# 4. Фильтруем города Центрального федерального округа
central_df = df[df['Федеральный округ'] == 'Центральный'].copy()
```

```python
# 5. Функция для расчёта расстояния (формула гаверсинуса)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Радиус Земли в километрах
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat / 2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c
```

```python
# 6. Добавляем колонку "Расстояние до Москвы"
central_df['Расстояние до Москвы'] = central_df.apply(
    lambda row: haversine(moscow_lat, moscow_lon, row['Широта'], row['Долгота']),
    axis=1
)
```

```python
# 8. Фильтруем города: расстояние <= 100 км и население <= 50 000
filtered = central_df[
    (central_df['Расстояние до Москвы'] <= 100) &
    (central_df['Население'] <= 50000)
]
```

```python
# 3. Выводим результат
filtered[['Город', 'Регион', 'Федеральный округ', 'Население', 'Расстояние до Москвы']]
```

```python
# 1. Ищем Новосибирск
novosibirsk_row = df[df['Город'] == 'Новосибирск'].iloc[0]
novosibirsk_lat = novosibirsk_row['Широта']
novosibirsk_lon = novosibirsk_row['Долгота']
```

```python
# 2. Оставляем только города из Сибирского федерального округа
siberian_df = df[df['Федеральный округ'] == 'Сибирский'].copy()
```

```python
# 6. Выводим результат
filtered[['Город', 'Регион', 'Федеральный округ', 'Население', 'Расстояние до Новосибирска']]
```

3. Найдите город(а), которые расположены в Дальневосточном федеральном округе и ближе всего к административному центру(Владивосток) с населением от 10т. до 20т. человек (5 городов)

```python
# 1. Ищем Владивосток
vladivostok_row = df[df['Город'] == 'Владивосток'].iloc[0]
vladivostok_lat = vladivostok_row['Широта']
vladivostok_lon = vladivostok_row['Долгота']
```

```python
# 2. Оставляем только города из Дальневосточного федерального округа
fareast_df = df[df['Федеральный округ'] == 'Дальневосточный'].copy()
```

```python
# 4. Добавляем колонку "Расстояние до Владивостока"
fareast_df['Расстояние до Владивостока'] = fareast_df.apply(
    lambda row: haversine(vladivostok_lat, vladivostok_lon, row['Широта'], row['Долгота']),
    axis=1
)
```

```python
# 5. Фильтруем по условиям
filtered = fareast_df[
    (fareast_df['Население'] >= 10000) &
    (fareast_df['Население'] <= 20000)
]
```

```python
# 6. Сортируем по расстоянию и берём 5 ближайших
top_5 = filtered.sort_values(by='Расстояние до Владивостока').head(5)
top_5[['Город','Население', 'Расстояние до Владивостока']]
```

4. Найдите три самых многочисленных округа с радиусом 100 км от административного центра, не считая ЦФО.  
 Подсказка для .query: имена с пробелами можно заключить в `имя с пробелом`

```python
# 1. Посмотреть все уникальные значения в столбце
df['Признак центра района или региона'].unique()
```

```python
# 2. Перебираем все уникальные флаги
for flag in sorted(df['Признак центра района или региона'].unique()):
    print(f"\n--- Примеры для флага {flag} ---")
    display(df[df['Признак центра района или региона'] == flag][['Город', 'Регион', 'Федеральный округ']].head(5))
```

```python
# 3. Убираем Центральный федеральный округ
df_no_cfo = df[df['Федеральный округ'] != 'Центральный'].copy()
```

```python
# 5. Берём уникальные округа (без Центрального)
districts = df_no_cfo['Федеральный округ'].unique()
```

```python
# 6. Проходим по каждому федеральному округу
result = {}
for district in districts:
    district_cities = df_no_cfo[df_no_cfo['Федеральный округ'] == district].copy()
    
    # Находим административный центр (флаг 2 или 3)
    capital_row = district_cities[district_cities['Признак центра района или региона'].isin([2, 3])]
    
    if capital_row.empty:
        continue  # если нет центра — пропускаем округ
    
    # Берем первую найденную столицу
    capital_row = capital_row.iloc[0]
    capital_lat = capital_row['Широта']
    capital_lon = capital_row['Долгота']
    
    # Считаем расстояния до всех городов округа
    district_cities['Расстояние до центра'] = district_cities.apply(
        lambda row: haversine(capital_lat, capital_lon, row['Широта'], row['Долгота']),
        axis=1
    )
    
    # Оставляем города в радиусе 100 км
    nearby_cities = district_cities.query('`Расстояние до центра` <= 100')
    
    # Считаем общее население
    total_population = nearby_cities['Население'].sum()
    
    # Записываем результат
    result[district] = total_population
```

```python
# 7. Переводим результат в датафрейм
result_df = pd.DataFrame(result.items(), columns=['Федеральный округ', 'Суммарное население'])
```

```python
# 8. Сортируем по убыванию и берём топ-3 округа
top_3 = result_df.sort_values(by='Суммарное население', ascending=False).head(3)

top_3
```
