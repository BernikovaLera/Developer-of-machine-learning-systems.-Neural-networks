## Задание

  
Используя датасет `cities.csv`, выполните следующие задания:

1. Найдите города, которые расположены в Центральном федеральном округе в радиусе 100 км от административного центра(Москва) с населением <= 50т. человек  

  
2. Найдите города, которые расположены в Сибирском федеральном округе в радиусе от 150 до 250 км от административного центра(Новосибирск) с населением  от 100т. до 200т. человек  
  
  
3. Найдите город(а), которые расположены в Дальневосточном федеральном округе  и ближе всего к административному центру(Владивосток) с населением  от 10т. до 20т. человек (5 городов)  
  
  
4. Найдите три самых многочисленных округа с радиусом 100 км от административного центра, не считая ЦФО.  
Подсказка для `.query`: имена с пробелами можно заключить в **\`имя с пробелом\`**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/cities.csv')
```

```python
df[["Город", 'Широта', 'Долгота']].query('Город == "Владивосток"')
```

```python
df[["Город", 'Широта', 'Долгота']].query('Город == "Новосибирск"')
```

```python
def haversine(lat1, lon1, lat2, lon2):
    RADIUS = 6371.2
    lat1, lon1, lat2, lon2 = map(np.deg2rad, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    total_km = RADIUS * c
    return total_km
```

```python
# Найдите города, которые расположены в Центральном федеральном округе в радиусе 100 км от административного центра(Москва) с населением <= 50т. человек
def haversine_looping(df):
    distance_list = []
    for i in range(len(df)):
        d = haversine(55.7522, 37.6156, df.iloc[i]['Широта'], df.iloc[i]['Долгота'])
        distance_list.append(d)
    return distance_list

df['Расстояние'] = haversine(55.7522, 37.6156, df['Широта'].values, df['Долгота'].values)
df['Расстояние'] = df['Расстояние'].round(2)

df1 = df[["Город",'Расстояние','Население']]
df1 = df1.dropna()
df1['Население'] = pd.to_numeric(df['Население'], errors = 'coerce')
df1 = df1[(df1 ['Расстояние'] < 100) & (df1 ['Население'] <= 50000)]

df1[["Город",'Расстояние','Население']]
```

```python
# Найдите города, которые расположены в Сибирском федеральном округе в радиусе от 150 до 250 км от административного центра(Новосибирск) с населением от 100т. до 200т. человек

def haversine_looping(df):
    distance_list = []
    for i in range(len(df)):
        d = haversine(55.028102, 82.921058, df.iloc[i]['Широта'], df.iloc[i]['Долгота'])
        distance_list.append(d)
    return distance_list

df['Расстояние'] = haversine(55.028102, 82.921058, df['Широта'].values, df['Долгота'].values)
df['Расстояние'] = df['Расстояние'].round(2)

df1 = df[["Город",'Расстояние','Население','Федеральный округ']]
df1 = df1.dropna()
df1['Население'] = pd.to_numeric(df['Население'], errors = 'coerce')
df1 = df1[
    (df1 ['Расстояние'] < 250) &
    (df1 ['Расстояние'] > 150) &
    (df1 ['Население'] >= 100_000) &
    (df1 ['Население'] <= 200_000) &
    (df1 ['Федеральный округ'] == 'Сибирский')
]

df1[["Город",'Расстояние','Население', 'Федеральный округ']].head()
```

```python
# Найдите город(а), которые расположены в Дальневосточном федеральном округе и ближе всего к административному центру(Владивосток) с населением от 10т. до 20т. человек (5 городов)
def haversine_looping(df):
    distance_list = []
    for i in range(len(df)):
        d = haversine(43.116381, 131.882348, df.iloc[i]['Широта'], df.iloc[i]['Долгота'])
        distance_list.append(d)
    return distance_list

df['Расстояние'] = haversine(43.116381, 131.882348, df['Широта'].values, df['Долгота'].values)
df['Расстояние'] = df['Расстояние'].round(2)

df1 = df[["Город",'Расстояние','Население','Федеральный округ']]
df1 = df1.dropna()
df1['Население'] = pd.to_numeric(df['Население'], errors = 'coerce')
df1 = df1[
    (df1 ['Расстояние'] > 0) &
    (df1 ['Население'] >= 10_000) &
    (df1 ['Население'] <= 20_000) &
    (df1 ['Федеральный округ'] == 'Дальневосточный')
]
sorted_df = df1.sort_values (by=['Расстояние'], ascending=[True])

sorted_df [["Город",'Расстояние','Население', 'Федеральный округ']].head()
```
