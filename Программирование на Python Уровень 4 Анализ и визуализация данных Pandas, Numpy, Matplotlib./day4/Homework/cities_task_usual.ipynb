## Задание

  
Используя датасет `cities.csv`, выполните следующие задания:

1. Найдите города, которые расположены в Центральном федеральном округе в радиусе 100 км от административного центра(Москва) с населением <= 50т. человек  

  
2. Найдите города, которые расположены в Сибирском федеральном округе в радиусе от 150 до 250 км от административного центра(Новосибирск) с населением  от 100т. до 200т. человек  


```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
def haversine(lat1, lon1, lat2, lon2):
    RADIUS = 6371.2
    lat1, lon1, lat2, lon2 = map(np.deg2rad, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    total_km = RADIUS * c
    return total_km
```

```python
data = pd.read_csv('data/cities.csv')
data['Население'] = pd.to_numeric(data['Население'], errors='coerce')
```

```python
data_tsentr=data[data['Федеральный округ'] == 'Центральный']

display(data_tsentr.loc[
    (data_tsentr['Федеральный округ'] == 'Центральный') &
    (haversine(55.7522, 37.6156,
               data_tsentr['Широта'],
               data_tsentr['Долгота']) <= 100
    ) & (data_tsentr['Население'] <= 50000), ['Город']])
```

```python
data_sib = data[data['Федеральный округ'] == 'Сибирский']

display(data_sib.loc[
    ((cur_res := haversine(55.028102, 82.921058, data_sib['Широта'], data_sib['Долгота'])) >= 150) &
    (cur_res <= 250) &
    (data_sib['Население'] >= 100_000) & (data_sib['Население'] <= 200_000),
    ['Город']]
)
```
