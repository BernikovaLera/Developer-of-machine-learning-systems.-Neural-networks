#### Задача 1

**1.1** Создать массив 10 x 10 со случайными значениями от 1 до 100(int64)

**1.2** Вычислить среднее значение всех элементов массива с помощью `np.mean()`

**1.3** Вычислить среднее значение всех элементов для каждой строки с помощью `np.mean()` и добавить как новый столбец

**1.4** Вычислить сумму значений всех элементов для каждого столбца с помощью `np.sum()`и добавить как новую строку, учитывая результат пункта `1.3`

```python
mas = np.random.randint(1, 100, size=(10, 10))
print(mas)
avg_all = np.mean(mas)
print(avg_all)
avg1 = np.mean(mas, axis=1)
print(avg1)
mas = mas.transpose()
mas = np.append(mas, [avg1], axis = 0)
mas = mas.transpose()
print(mas)
sm = np.sum(mas, axis=0)
print(sm)
mas = np.append(mas, [sm], axis=0)
print(mas)
```

#### Задача 2
**2.1** Создать матрицу 7x7 с значениями `1,2,3,4,5,6,7` по главной диагонали, остальные нули. 
Два варианта решения: с помощью функции `np.diag()` и без(используя _broadcasting_).

```python
mas = np.arange(7) + 1
mas = np.diag(mas)
print(mas)
mas = np.eye(7, dtype=int)
mas[1,1] = 2
mas[2,2] = 3
mas[3,3] = 4
mas[4,4] = 5
mas[5,5] = 6
mas[6,6] = 7
print(mas)
```

#### Задача 3

**3.1** Создать случайную матрицу 5x7 целых чисел от 1 до 100 и поменять местами 3-ю (индекс 2) и 5-ю (индекс 4) колонки

```python
mas = np.random.randint(1, 100, size=(5,7))
print(mas)
col3 = [mas[0, 2], mas[1, 2], mas[2, 2], mas[3, 2], mas[4, 2]]
print(col3)
col5 = [mas[0, 4], mas[1, 4], mas[2, 4], mas[3, 4], mas[4, 4]]
print(col5)
mas[0,2] = col5[0]
mas[1,2] = col5[1]
mas[2,2] = col5[2]
mas[3,2] = col5[3]
mas[4,2] = col5[4]
mas[0,4] = col3[0]
mas[1,4] = col3[1]
mas[2,4] = col3[2]
mas[3,4] = col3[3]
mas[4,4] = col3[4]
print(mas)
```
